```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
BEC = pd.read_csv('GSE63142_BAL_out.txt', header=None, sep=' ')
```

```python
BEC
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(.95)
```

```python
pca.fit(BEC)
```

```python
transformed_data = pca.transform(BEC)
```

```python
pca_BEC = pd.DataFrame(transformed_data)
```

```python
pca_BEC.head()
```

```python
BAL = pd.read_csv('GSE67940_BEC_out.txt', header = None, sep=' ')
```

```python
BAL
```

```python
pca.fit(BAL)
```

```python
transformed_data = pca.transform(BAL)
```

```python
pca_BAL = pd.DataFrame(transformed_data)
```

```python
pca_BAL.head()
```

```python
pca_BEC.to_pickle("./pca_BEC.pkl")
```

```python
pca_BAL.to_pickle("./pca_BAL.pkl")
```

```python
pca = PCA(n_components=2)
components = pca.fit_transform(BAL)
```
